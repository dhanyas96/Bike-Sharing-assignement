```python
# Generic function to calculate VIF of variables

def VIF_Calculation(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# calculate VIF

VIF_Calculation(X_train_rfe)
```

Humidity shows a very high VIF value of 30.89.

1. X and Y should display some sort of a linear relationship; otherwise, there is no use of fitting a linear model between them.
